# Models d'aprenentatge supervisat

Quan tenim un dataset etiquetat en el que un dels atributs és el `target` o `label` i es tracta de predir `tagets` de noves instàncies. 

Dins d'aquests els dos tipus de problemes són `classificació` si l'atribut `target` és categòric i `Regressió` si és numèric.

Tots els problemes plantejen una funció que té uns `paràmetres`. A la funció li arriben els `atributs predictors`, li aplica els paràmetres i trau una predicció. Els paràmetres i la funció defineixen el model. Com que la funció sempre serà igual el que cal guardar són els paràmetres. El que s'entrena són els valors dels paràmetres per minimitzar l'error de la predicció respecte als valors reals.
